```python
import torch
import torch.nn as nn

class GeneCNNLSTMModel(nn.Module):
    def __init__(self, num_genes, embedding_dim, num_classes, cnn_filters=64, lstm_hidden_size=128, kernel_size=3, dropout_rate=0.5):
        super(GeneCNNLSTMModel, self).__init__()
        
        # 유전자 임베딩 레이어
        self.embedding = nn.Embedding(num_genes, embedding_dim)
        
        # CNN Layer 1
        self.conv1 = nn.Conv1d(in_channels=embedding_dim, out_channels=cnn_filters, kernel_size=int(kernel_size), padding=1)
        self.conv2 = nn.Conv1d(in_channels=cnn_filters, out_channels=cnn_filters*2, kernel_size=int(kernel_size), padding=1)
        
        # Pooling Layer (1D Max Pooling)
        self.pool = nn.MaxPool1d(kernel_size=2)
        
        # LSTM Layer
        self.lstm = nn.LSTM(input_size=cnn_filters*2, hidden_size=lstm_hidden_size, batch_first=True)
        
        # Dropout Layer
        self.dropout = nn.Dropout(dropout_rate)
        
        # Fully Connected Layer
        self.fc = nn.Sequential(
            nn.Linear(lstm_hidden_size, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        # x: (batch_size, num_genes)
        x = self.embedding(x)  # (batch_size, num_genes, embedding_dim)
        
        # 1D CNN 적용을 위해 차원 변환 (batch_size, embedding_dim, num_genes)
        x = x.permute(0, 2, 1)
        
        # 여러 CNN 레이어 적용
        x = self.conv1(x)
        x = self.pool(x)  # 크기 감소
        x = self.conv2(x)
        x = self.pool(x)  # 크기 감소 (batch_size, cnn_filters*2, num_genes // 4)
        
        # CNN 출력을 LSTM에 전달
        x = x.permute(0, 2, 1)  # (batch_size, num_genes // 4, cnn_filters*2)
        lstm_out, (h_n, c_n) = self.lstm(x)  # LSTM 출력 (batch_size, num_genes // 4, lstm_hidden_size)
        
        # LSTM 마지막 출력만 사용
        x = lstm_out[:, -1, :]  # (batch_size, lstm_hidden_size)
        
        # 드롭아웃 적용
        x = self.dropout(x)
        
        # Fully Connected Layer 적용
        out = self.fc(x)  # (batch_size, num_classes)
        
        return out

# 하이퍼파라미터 설정
num_genes = 274  # 유전자의 수
embedding_dim = 64  # 각 유전자를 표현할 임베딩 차원
num_classes = 26  # 암의 종류
cnn_filters = 64  # 1D CNN 필터 개수
lstm_hidden_size = 128  # LSTM 은닉층 크기
dropout_rate = 0.5  # 드롭아웃 비율
kernel_size = 3  # CNN 커널 크기

# 모델 생성
model = GeneCNNLSTMModel(num_genes, embedding_dim, num_classes, cnn_filters, lstm_hidden_size, kernel_size, dropout_rate)
```

```python
import torch
import torch.nn as nn

# Attention 메커니즘 정의
class Attention(nn.Module):
    def __init__(self, input_dim):
        super(Attention, self).__init__()
        self.attention = nn.Sequential(
            nn.Linear(input_dim, 128),
            nn.ReLU(),
            nn.Linear(128, 1)
        )

    def forward(self, x):
        # x: (batch_size, seq_len, input_dim)
        weights = self.attention(x)  # (batch_size, seq_len, 1)
        weights = torch.softmax(weights, dim=1)  # Attention weights 계산 (batch_size, seq_len, 1)
        weighted_output = x * weights  # 가중합 적용 (batch_size, seq_len, input_dim)
        return weighted_output.sum(dim=1)  # Attention을 통해 가중합된 출력 (batch_size, input_dim)

# CNN + Attention 모델 정의
class GeneCNNAttentionModel(nn.Module):
    def __init__(self, num_genes, embedding_dim, num_classes, cnn_filters=64, kernel_size=3, dropout_rate=0.5):
        super(GeneCNNAttentionModel, self).__init__()
        
        # 유전자 임베딩 레이어
        self.embedding = nn.Embedding(num_genes, embedding_dim)
        
        # CNN Layer
        self.conv1 = nn.Conv1d(in_channels=embedding_dim, out_channels=cnn_filters, kernel_size=kernel_size, padding=1)
        self.conv2 = nn.Conv1d(in_channels=cnn_filters, out_channels=cnn_filters*2, kernel_size=kernel_size, padding=1)
        
        # Pooling Layer (1D Max Pooling)
        self.pool = nn.MaxPool1d(kernel_size=2)
        
        # Attention 메커니즘
        self.attention = Attention(cnn_filters*2)
        
        # Dropout Layer
        self.dropout = nn.Dropout(dropout_rate)
        
        # Fully Connected Layer
        self.fc = nn.Sequential(
            nn.Linear(cnn_filters*2, 128),
            nn.ReLU(),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        # x: (batch_size, num_genes)
        x = self.embedding(x)  # (batch_size, num_genes, embedding_dim)
        
        # 1D CNN 적용을 위해 차원 변환 (batch_size, embedding_dim, num_genes)
        x = x.permute(0, 2, 1)
        
        # CNN 적용
        x = self.conv1(x)
        x = self.pool(x)  # (batch_size, cnn_filters, num_genes // 2)
        x = self.conv2(x)
        x = self.pool(x)  # (batch_size, cnn_filters*2, num_genes // 4)
        
        # CNN 출력을 Attention 메커니즘에 전달
        x = x.permute(0, 2, 1)  # (batch_size, num_genes // 4, cnn_filters*2)
        x = self.attention(x)  # (batch_size, cnn_filters*2)
        
        # 드롭아웃 적용
        x = self.dropout(x)
        
        # Fully Connected Layer 적용
        out = self.fc(x)  # (batch_size, num_classes)
        
        return out

# 하이퍼파라미터 설정
num_genes = 274  # 유전자의 수
embedding_dim = 64  # 각 유전자를 표현할 임베딩 차원
num_classes = 26  # 암의 종류
cnn_filters = 64  # 1D CNN 필터 개수
kernel_size = 3  # CNN 커널 크기
dropout_rate = 0.5  # 드롭아웃 비율

# 모델 생성
model = GeneCNNAttentionModel(num_genes, embedding_dim, num_classes, cnn_filters, kernel_size, dropout_rate)
```

```python
import torch
import torch.nn as nn
from sklearn.metrics import f1_score, accuracy_score

# 학습 및 평가 함수
def train_and_evaluate(model, train_loader, val_loader, criterion, optimizer, num_epochs=25, patience=20, device='cpu'):
    model.to(device)  # 모델을 GPU/CPU로 이동
    best_val_f1 = 0.0  # 검증에서 가장 높은 F1-score 기록
    epochs_no_improve = 0  # 개선되지 않은 에포크 수 기록

    # 학습률 스케줄러 정의
    scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)
    
    for epoch in range(num_epochs):
        # 모델을 학습 모드로 설정
        model.train()
        running_loss = 0.0
        
        for batch_X, batch_y in train_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            
            optimizer.zero_grad()  # 경사 초기화
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            
            running_loss += loss.item()

        # 검증 단계
        val_f1, val_acc, val_loss = evaluate(model, val_loader, criterion, device)
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}, '
              f'Val Loss: {val_loss:.4f}, Val F1: {val_f1:.4f}, Val Accuracy: {val_acc:.4f}')
        
        # 스케줄러 업데이트
        scheduler.step()

        # Early Stopping 로직 적용
        if val_f1 > best_val_f1:
            best_val_f1 = val_f1
            torch.save(model.state_dict(), 'best_model.pth')  # 최적 모델 저장
            epochs_no_improve = 0  # 성능이 개선되었으므로 초기화
        else:
            epochs_no_improve += 1

        # patience만큼 에포크 동안 개선이 없으면 조기 종료
        if epochs_no_improve >= patience:
            print(f'Early stopping at epoch {epoch+1}')
            break

    print(f'Best Validation F1 Score: {best_val_f1:.4f}')

# 검증 단계 함수
def evaluate(model, val_loader, criterion, device='cpu'):
    model.eval()  # 평가 모드로 설정
    all_preds = []
    all_labels = []
    total_loss = 0.0
    
    with torch.no_grad():
        for val_X, val_y in val_loader:
            val_X, val_y = val_X.to(device), val_y.to(device)
            outputs = model(val_X)
            loss = criterion(outputs, val_y)
            total_loss += loss.item()

            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(val_y.cpu().numpy())

    # F1-score 및 Accuracy 계산
    val_f1 = f1_score(all_labels, all_preds, average='macro')
    val_acc = accuracy_score(all_labels, all_preds)
    
    return val_f1, val_acc, total_loss / len(val_loader)

import torch
import torch.nn as nn
import torch.nn.functional as F

class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, weight=None, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.weight = weight
        self.reduction = reduction

    def forward(self, inputs, targets):
        # CrossEntropyLoss와 유사한 형태로 예측값을 소프트맥스함
        CE_loss = F.cross_entropy(inputs, targets, weight=self.weight, reduction='none')
        
        # 각 샘플의 확률 값
        pt = torch.exp(-CE_loss)
        
        # Focal Loss 계산
        focal_loss = self.alpha * (1 - pt) ** self.gamma * CE_loss
        
        if self.reduction == 'mean':
            return torch.mean(focal_loss)
        elif self.reduction == 'sum':
            return torch.sum(focal_loss)
        else:
            return focal_loss

# 모델, 손실 함수, 옵티마이저 정의
#class_counts = np.bincount(y_train_resampled)
#class_weights = 1. / class_counts
#weights = torch.tensor(class_weights, dtype=torch.float32)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.0008)

# 모델 학습 및 평가 실행
train_and_evaluate(model, train_loader, val_loader, criterion, optimizer, num_epochs=500)
```
